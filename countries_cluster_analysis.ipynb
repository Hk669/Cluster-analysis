## IMPORT THE LIBs

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
kmeans = KMeans(7)
```

### LOAD DATA

```python
data = pd.read_csv("D:\Cluster analysis\Countries_cluster_analysis.csv")
data
```

```python
plt.scatter(data['Longitude'],data['Latitude'])
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.show()
```

```python
x = data.iloc[:,1:3]
x
```

### CLUSTER ANALYSIS

```python
cluster_column = kmeans.fit_predict(x)
cluster_column
```

```python
data_cluster = data.copy()
data_cluster['Cluster'] = cluster_column
data_cluster
```

### PLOT THE CLUSTERS

```python
plt.scatter(data['Longitude'],data['Latitude'],c=data_cluster['Cluster'],cmap='rainbow')
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.show()
```
